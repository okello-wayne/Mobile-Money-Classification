### Modelling phase

```python
#Separate training features from target
X_train = train_data.drop(['ID','Q1', 'Latitude', 'Longitude',
                           'savings','borrowing','mobile_money','insurance','mobile_money_classification'], axis=1)
y_train = train_data['mobile_money_classification']

X_test = test_data.drop(['ID','Q1', 'Latitude', 'Longitude'], axis=1)
#rescale X_train and X_test
# import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
X_train_scaled[:1]
```

```python
# show the shape after scalling 

print("Train set: {}".format(X_train_scaled.shape))
print("Test set: {}".format(X_test_scaled.shape)) 
```

```python
# Split train_data
from sklearn.model_selection import train_test_split

X_Train, X_val, y_Train, y_val = train_test_split(X_train_scaled, y_train, stratify = y_train,
                                                  test_size = 0.2, random_state=42)
```

```python
#import classifier algorithm here 
from sklearn.linear_model import LogisticRegression

#create your model
my_model = LogisticRegression()

#fitting the model
my_model.fit(X_Train,y_Train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

# fit by setting best parameters and Evaluate model
my_model = LogisticRegression()

my_model.fit(X_Train, y_Train)
y_pred = my_model.predict(X_val)

# Get error rate
print("Error rate of The Logistic Regression classifier: ", 1 - accuracy_score(y_val, y_pred))

# Get confusion matrix
confusion_matrix(y_pred, y_val)
```

```python
#Get log loss
from sklearn.metrics import log_loss

my_model = LogisticRegression()
my_model.fit(X_Train, y_Train)
pred = my_model.predict_proba(X_val)
eval = log_loss(y_val,pred) 
print('Log Loss is:',eval)
```

```python
set_pred = my_model.predict_proba(X_test_scaled)
pred = pd.DataFrame(set_pred, columns = ['no_finacial_services','other_only','mm_only','mm_plus'])
```

```python
org = test_data['ID']
subm = pd.concat([org,pred], axis=1)
subm.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
cf = RandomForestClassifier(n_estimators=100)
cf.fit(X_Train,y_Train)
y_pred = cf.predict(X_val)
print("Error rate of The Random Forrest classifier: ", 1 - accuracy_score(y_val, y_pred))
```

```python
pred = cf.predict_proba(X_val)
eval = log_loss(y_val,pred) 
print('Log Loss is:',eval)
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(15,), random_state=1)
clf.fit(X_Train,y_Train)
y_pred = clf.predict(X_val)
print("Error rate of The Neural Network classifier: ", 1 - accuracy_score(y_val, y_pred))
```

```python
pred = clf.predict_proba(X_val)
eval = log_loss(y_val,pred) 
print('Log Loss is:',eval)
```
